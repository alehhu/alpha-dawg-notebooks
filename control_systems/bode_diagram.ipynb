# Bode Diagram Analysis of a Transfer Function üìà

In this notebook, we analyze a linear time-invariant (LTI) transfer function using the `control` library.  
The goal is to study the system's frequency response through the Bode diagram, with observations and qualitative analysis.

## üîç Objective
Study the transfer function:

$$
G(s) = \frac{10}{s^2 + 2s + 10}
$$

and make observations on gain, phase, stability, and dynamic behavior.

```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# Define the transfer function
num = [10]  # Numerator
den = [1, 2, 10]  # Denominator
G = ctrl.TransferFunction(num, den)

# Generate a logarithmic frequency range
omega = np.logspace(-1, 2, 100)

# Calculate frequency response
mag, phase, omega = ctrl.frequency_response(G, omega)

# Plot the Bode diagram
plt.figure()
plt.subplot(2, 1, 1)
plt.semilogx(omega, 20 * np.log10(mag))  # Magnitude in dB
plt.title('Bode Diagram')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Magnitude [dB]')

plt.subplot(2, 1, 2)
plt.semilogx(omega, phase * 180 / np.pi)  # Phase in degrees
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Phase [degrees]')

plt.tight_layout()
plt.show()
```

## üìä Observations from the Bode Diagram

- **Low-frequency gain**: The gain is constant around ~20 dB.
- **Resonance frequency**: A slight peak is observed around ~3 rad/s.
- **Phase**: The phase starts from 0¬∞ and gradually decreases toward ~-180¬∞.

The system shows typical behavior of a **slightly underdamped second-order system**:
- The peak indicates mild resonance.
- The gain slope beyond the cutoff frequency tends to -40 dB/decade.
- The phase suggests a complex conjugate pair of poles.

---

üìå *Technical note*: The cutoff frequency can be estimated as the frequency where the magnitude drops by 3 dB from its initial value.

## ‚úÖ Conclusions

The system is **stable**, and the frequency response suggests good dynamic behavior for applications with moderate speed requirements.  
However, the resonance peak could cause overshoot in response to rapid signals.
